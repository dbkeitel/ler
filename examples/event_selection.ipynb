# Event selection

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline  
from sklearn.neighbors import KernelDensity
import ler
from ler import LeR
```

```python
# callin class
ler_ = LeR(npool=6,z_min=0., z_max=10.,snr_type = 'interpolation',\
           psds={'L1':'aLIGOaLIGODesignSensitivityT1800044','H1':'aLIGOaLIGODesignSensitivityT1800044','V1':'AdvVirgo'}, \
           waveform_inspiral_must_be_above_fmin=True)
```

```python
ler_.selecting_n_detectable_events?
```

# 4 image lensed: 2 detectable Type I 

```python
# suprathreshold case
param1 = ler_.selecting_n_detectable_events(lenstype='I',snr_threshold=8.0, num_img=2, new=True, min_img=4, size=1000, batch_size=10000)
```

# 4 image lensed: 2 detectable Type II

```python
# suprathreshold case
param2 = ler_.selecting_n_detectable_events(lenstype='II',snr_threshold=8.0, num_img=2, new=True, min_img=4, size=1000, batch_size=10000)
```

```python
# magnification
# type I
amp = param1['magnifications']
amp1_1 = amp[:,0]
amp1_2 = amp[:,1]
dmu12 = abs(amp1_2-amp1_1)
# type II
amp = param2['magnifications']
amp2_1 = amp[:,2]
amp2_2 = amp[:,3]
dmu23 = abs(amp2_2-amp1_1)

# time delays
# type I
dt = param1['time_delays']/(60*60*24) #
delay1_1 = dt[:,0]
delay1_2 = dt[:,1]
dt12 = abs(delay1_2-delay1_1)
# type II
dt = param2['time_delays']/(60*60*24)
delay2_1 = dt[:,2]
delay2_2 = dt[:,3]
dt23 = abs(delay2_2-delay2_1)
```

```python
min_ = -4
max_ = 4
# type I
x12 = np.log10(dt12)
kde = KernelDensity(bandwidth=1, kernel='gaussian')
kde.fit(x12[:, np.newaxis])
x_d = np.linspace(min_, max_, 1000)
logprob = kde.score_samples(x_d[:, np.newaxis])
plt.fill_between(10**x_d, np.exp(logprob), alpha=0.5)
plt.plot(10**x_d, np.exp(logprob), '-')
# type II
x23 = np.log10(dt23)
kde = KernelDensity(bandwidth=1, kernel='gaussian')
kde.fit(x23[:, np.newaxis])
x_d = np.linspace(min_, max_, 1000)
logprob = kde.score_samples(x_d[:, np.newaxis])
plt.fill_between(10**x_d, np.exp(logprob), alpha=0.5)
plt.plot(10**x_d, np.exp(logprob), '-')

plt.xscale('log')
plt.ylabel('pdf')
plt.xlabel('days')
plt.title('Time delay between same image type in 4 image events')
plt.show()
```

```python
dt  = param1['time_delays']
image = np.log10(np.array([dt[:,1],dt[:,2],dt[:,3]])/(24*60*60))

for i in range(len(image)):
    x_train = image[i] 
    kde = KernelDensity(bandwidth=1, kernel='gaussian')
    kde.fit(x_train[:, np.newaxis])
    min_ = -4
    max_ = 4
    x_d = np.linspace(min_, max_, 1000)
    logprob = kde.score_samples(x_d[:, np.newaxis])
    
    name = 'Image '+ str(i+2)
    #plt.fill_between(x_d, np.exp(logprob), alpha=0.3)
    plt.plot(10**x_d, np.exp(logprob), '-', alpha=0.4, label=name)
    
x_train = np.log10(np.concatenate((dt[:,1],dt[:,2],dt[:,3]))/(24*60*60))
kde = KernelDensity(bandwidth=1, kernel='gaussian')
kde.fit(x_train[:, np.newaxis])
min_ = -4
max_ = 4
x_d = np.linspace(min_, max_, 1000)
logprob = kde.score_samples(x_d[:, np.newaxis])

name = 'Image all'
#plt.fill_between(x_d, np.exp(logprob), alpha=0.3)
plt.plot(10**x_d, np.exp(logprob), '-', alpha=0.8, label=name)
    
#plt.xlim(0.00001,5)
#plt.ylim(0,0.2)
plt.xscale('log')
plt.ylabel('pdf')
plt.xlabel('days')
plt.title('Time delay wrt to 1st image')
plt.legend()
plt.show()
```
