```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from ler import SourceGalaxyPopulationModel
```

```python
# default model
# BBH popI/II Oguri
cbc = SourceGalaxyPopulationModel(merger_rate_density="merger_rate_density_bbh_popI_II_oguri")
```

```python
z = np.linspace(0, 10, 100)
dVc_dz = cbc.differential_comoving_volume(z)
plt.scatter(z, dVc_dz)
plt.show()
```

```python
# source frame
z = np.linspace(0, 10, 1000)
rate = (1/cbc.normalization_pdf_z) * cbc.merger_rate_density(z)/(1+z) * cbc.differential_comoving_volume(z)
```

```python
# plot the merger rate density
plt.scatter(z, rate)
```

```python
# pop III
cbc = SourceGalaxyPopulationModel(z_min=0.0,
        z_max=40.0,
        merger_rate_density="merger_rate_density_popIII_ken")

# source frame
z = np.linspace(5, 40, 1000)
rate = cbc.merger_rate_density(z)
#rate = (1/cbc.normalization_pdf_z) * cbc.merger_rate_density(z)/(1+z) * cbc.differential_comoving_volume(z)
# plot the merger rate density
plt.scatter(z, rate)
# y-axis in log scale
plt.yscale('log')
```
