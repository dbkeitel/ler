```python
import numpy as np
import matplotlib.pyplot as plt
from lenstronomy.LensModel.lens_model import LensModel
from lenstronomy.Plots.lens_plot import lens_model_plot
from ipywidgets import interactive
```

```python
# Define the lens model
lens_model = LensModel(lens_model_list=['POINT_MASS'])
kwargs_lens = [{'theta_E': 1.0, 'center_x': 0.0, 'center_y': 0.0}]  # Einstein radius and lens position

def plot_lens_system(beta_x, beta_y):
    """
    Plots the lens system for given source positions.
    """
    # Source position
    kwargs_source = {'center_x': beta_x, 'center_y': beta_y}
    
    # Generate lensing quantities
    f, ax = plt.subplots(1, 1, figsize=(8, 8))
    lens_model_plot(ax, lensModel=lens_model, kwargs_lens=kwargs_lens, sourcePos_x=beta_x, sourcePos_y=beta_y,
                    with_caustics=True, point_source=True, numPix=100, deltaPix=0.05, 
                    source_amp=1000, no_frame=False, band_plot=True)
    
    plt.show()

# Create an interactive plot with sliders for the source position
interactive_plot = interactive(plot_lens_system, beta_x=(-2.0, 2.0, 0.1), beta_y=(-2.0, 2.0, 0.1))
interactive_plot
```
