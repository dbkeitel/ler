```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Constants
G = 6.67430e-11  # gravitational constant in m^3 kg^-1 s^-2
M = 5.972e24     # mass of the object (e.g., Earth) in kg
c = 299792458    # speed of light in m/s

# Schwarzschild radius
r_s = 2 * G * M / c**2

# Define a grid of points
x = np.linspace(-2*r_s, 2*r_s, 400)
y = np.linspace(-2*r_s, 2*r_s, 400)
x, y = np.meshgrid(x, y)
z = np.sqrt(np.abs(r_s**2 - x**2 - y**2))

# Mask for points outside the Schwarzschild radius
mask = np.isnan(z)
z[mask] = 0

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.plot_surface(x, y, z, color='b', alpha=0.5)
ax.plot_surface(x, y, -z, color='b', alpha=0.5)

ax.set_xlabel('X Coordinate')
ax.set_ylabel('Y Coordinate')
ax.set_zlabel('Z Coordinate')

plt.title('Visualization of Space Curvature by a Black Hole')
plt.show()
```
